```python
for (i,j) in spotify_df.iterrows():
    spotify_df.loc[i,"Top Genre"]=spotify_df.loc[i,"Top Genre"].split()[-1]
```

```python
spotify_df.info()
```

```python
df=spotify_df.groupby(['Top Genre'])["Top Genre"]
df2=pd.DataFrame(df.describe())
type(df2)
```

```python
df2
```

```python
df2=df2.sort_values("count",ascending=False)
```

```python
df2 = df2[df2["count"]>10]
df2 = df2[df2["count"]<100]
df2
```

```python
genre_list=list(df2.index)
genre_list
```

```python
cur_df = spotify_df[spotify_df["Top Genre"].isin(genre_list)]
```

```python
cur_df.info()
```

```python
features = ["Energy","Loudness (dB)","Speechiness","Acousticness","Valence"]
```

```python
from sklearn.model_selection import train_test_split
X=cur_df[features]
y=cur_df["Top Genre"]
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_scaled
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y_encoded = le.fit_transform(y)
y_encoded.reshape(y_encoded.shape[0])
```

```python
X_train,X_test,y_train,y_test = train_test_split(X_scaled,y_encoded,test_size=0.2,random_state=0)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
forest = RandomForestClassifier(n_estimators=10, random_state=0)
forest.fit(X_train, y_train)
forest.score(X_test, y_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
tree = DecisionTreeClassifier()
tree.fit(X_train, y_train)
tree.score(X_test, y_test)
```

```python
from sklearn.svm import SVC
svm = SVC(kernel='rbf')
svm.fit(X_train, y_train)
svm.score(X_test, y_test)
```

```python
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression()
log_reg.fit(X_train, y_train)
log_reg.score(X_test, y_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
knn.score(X_test, y_test)
```
